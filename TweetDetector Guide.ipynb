# How to Run this Program

- Select 'Run' on your compiler. 
- A prompt should then show on screen, where you can input values



- Input what hashtag you are looking to search by in the first box (CAN ONLY SEARCH BY ONE HASHTAG)
- Input the timeframe you are looking to search SINCE in the second box (CAN ONLY SEARCH BY ONE DATE)

Then, select the option which you are looking to use
- Scrape Posts (Scrapes Posts from Twitter According to your Hashtag)
- Scrape Images (Scrapes Posts from Twitter According to your Hashtag, ALONG with Image and Media Links)
- Show Output CSV Files (Showcase the CSV file you just created in the program)
- Detect Corpus (Creates another CSV file based on the corpus you establish from Scraped_Posts CSV file !ENSURE CORPUS IS ADJUSTED TO YOUR SEARCH PREFERENCES!)
- Detect Images (Detects Images in IMAGES folder and label them manually according to what is in the Image)
